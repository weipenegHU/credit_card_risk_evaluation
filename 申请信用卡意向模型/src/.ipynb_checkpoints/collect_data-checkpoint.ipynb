```python
import mysql.connector
from mysql.connector import Error
import pandas as pd
import re
import glob2
import numpy as np
import os
import shutil
```

```python
def peek(df, line=None):
    print(df.shape)
    if line is None:
        print(df.head())
    else:
        print(df.head(line))
```

```python
def create_db_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
                        host=host_name,
                        user=user_name,
                        passwd=user_password,
                        database=db_name)
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
```

```python
def read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        print("execute query successfully")
        return result
    except Error as err:
        print(f"Error: '{error}'")
        
```

```python
# 查找指定时间段内推送的模板消息
find_mobanxiaoxi = \
'''
SELECT a.exec_date_key, a.job_key, a.job_name,
             (CASE a.wechat_account
              WHEN 'mfykzs' THEN '魔方用卡助手'
                WHEN 'mfykxzs' THEN '魔方用卡小助手'
                WHEN 'mfbk' THEN '魔方办卡'
                WHEN 'mfekt' THEN '魔方E卡通'
                WHEN 'kkykzs' THEN '卡卡用卡助手'
                WHEN 'kkyzks' THEN '卡卡通卡助手'
                WHEN 'kywy' THEN '卡易无忧'
                WHEN 'kyzs' THEN '卡易助手'
                WHEN 'sldkyzj' THEN '刷立得卡银专家'
                WHEN 'zkwy' THEN '众疌卡无忧'
                WHEN 'mfsk' THEN '魔方申卡小助手'
                WHEN 'mfkzs' THEN '魔方卡助手'
                WHEN 'mfbkvip' THEN '魔方办卡VIP'
                WHEN 'rjzn' THEN '魔方智卡'
                END) wechat_account, 
             b.card_name, a.commit_success_count, b.apply_cnt
FROM 
    (SELECT * FROM dim_dmp_job job 
    LEFT JOIN dim_dmp_job_config config USING (job_key) 
    LEFT JOIN dim_dmp_send_record record USING (job_key)
    WHERE job.job_type_key = 1
    AND config.wechat_account IN ('mfykxzs','mfykzs','mfbk','kkykzs','kywy','kyzs',
                                  'mfekt','sldkyzj','zkwy','mfsk','mfkzs','mfbkvip','rjzn')
-- 指定收集时间段
    AND job.exec_date_key BETWEEN 20220228 AND 20220331
-- 	AND config.wechat_account = "mfykxzs"
    ) a
LEFT JOIN 
    (SELECT info.task_key, ord.card_name, COUNT(DISTINCT ord.crm_order_key) 'apply_cnt' FROM fact_crm_order_activity_info info 
    LEFT JOIN fact_crm_order ord ON info.crm_order_key = ord.crm_order_key
    WHERE ord.product_type_key = 1
    GROUP BY info.task_key, ord.card_name
    ) b
ON a.job_key = b.task_key
WHERE b.card_name IS NOT NULL
ORDER BY exec_date_key
'''

connection = create_db_connection("am-wz98vqjuhh95217b1167320o.ads.aliyuncs.com","dmc_ads_read","yumbyb3JspVUvTHk", "dmc_abs_dw")
results = read_query(connection, find_mobanxiaoxi)
```

```python
# 整理模板消息收集结果
results = pd.DataFrame(results)
results.columns = ['exec_date_key','job_key','job_name','wechat_account','card_name','send_cnt','apply_cnt']
# 去除申请量较少的推送
results = results.query('(apply_cnt > 10) & (send_cnt > 0)')

results['apply_percent'] = results['apply_cnt'] / results['send_cnt']

# 去除响应率较低的推送, 因为这些推送很有可能是有问题的
use_results = results.query('apply_percent >= 0.01').drop(columns='apply_percent')
peek(use_results, 2)
```

```python
# 收集数据
use_results = use_results.reset_index(drop=True)
fail_job = []
# 设定数据集输出路径
outpath = ""
for i in range(use_results.shape[0]):
    exec_date_key, job_key, job_name, wechat_account, bank, _, _ = list(use_results.loc[i])
    job_name = re.sub(r"/",r"-", job_name)
    print(exec_date_key, job_key, wechat_account, bank)
    collect_train_data = \
    f'''
   WITH cte AS (
        SELECT ord.customer_key FROM fact_crm_order_activity_info info
        LEFT JOIN fact_crm_order ord ON ord.crm_order_key = info.crm_order_key
        WHERE info.task_key = {job_key}
    ),
    cte2 AS (
        SELECT customer_key FROM fact_dmp_send_record_detail WHERE job_key = {job_key} AND send_status_key <> 3
    )
    SELECT DISTINCT
                 a.*,
                 b.*,  
                 c.信用卡申请数,
                 c.信贷申请数,
                 c.核卡数,
                 c.被拒数,
                 c.核卡率,
                 c.被拒率,
                 d.最近一次申请产品,
                 e.线上申请信用卡次数,
                 e.线上申请信贷次数,
                 f.gap '距离上一次申请信用卡天数',
                 g.是否曾经申请过相同银行,
                 h.是否有该银行的核卡,
                 i.是否曾被该银行拒绝过,
                 J.是否和上一次申请的银行一致,
                 k.近一月推送次数
    FROM (

    (SELECT customer_key, 下单用户 FROM ( 
    SELECT *, IF(customer_key IN (SELECT customer_key FROM cte), 1, 0) '下单用户' FROM fact_dmp_send_record_detail detail
    WHERE send_status_key <> 3
    AND job_key = {job_key})) a

    LEFT JOIN

    (SELECT cus.customer_key, attr.age, attr.gender, DATEDIFF({exec_date_key}, cus.register_date_key) 'mob'
    -- , fattr.credit_card_apply_count, fattr.loan_apply_count, fattr.credit_card_audit_count
    FROM dim_customer cus
    LEFT JOIN dim_customer_common_attr attr ON cus.customer_key = attr.customer_key
    LEFT JOIN dim_customer_wechat wechat ON cus.customer_key = wechat.customer_key
    WHERE cus.customer_key IN (SELECT customer_key FROM cte2)
    ) b ON a.customer_key = b.customer_key
    -- LEFT JOIN dim_customer_finance_attr fattr ON cus.customer_key = fattr.customer_key
    LEFT JOIN
    -- 总的申请，审核
    (SELECT customer_key,  COUNT(DISTINCT IF(product_type_key = 1, ord.crm_order_key, NULL)) '信用卡申请数', COUNT(DISTINCT IF(product_type_key = 2, ord.crm_order_key, NULL)) '信贷申请数',
                 COUNT(DISTINCT IF(product_type_key = 1 AND order_status = "已核卡" AND audit_date_key < {exec_date_key}, ord.crm_order_key, NULL)) '核卡数', 
                 COUNT(DISTINCT IF(product_type_key = 1 AND order_status = "已拒绝" AND audit_date_key < {exec_date_key}, ord.crm_order_key, NULL)) '被拒数',
                 COUNT(DISTINCT IF(product_type_key = 1 AND order_status = "已核卡" AND audit_date_key < {exec_date_key}, ord.crm_order_key, NULL)) /  COUNT(DISTINCT IF(product_type_key = 1, ord.crm_order_key, NULL)) '核卡率',
                 COUNT(DISTINCT IF(product_type_key = 1 AND order_status = "已拒绝" AND audit_date_key < {exec_date_key}, ord.crm_order_key, NULL)) /  COUNT(DISTINCT IF(product_type_key = 1, ord.crm_order_key, NULL)) '被拒率'
    FROM fact_crm_order ord 
    WHERE ord.create_date_key < {exec_date_key}
    AND ord.product_type_key IN (1,2)
    AND customer_key IN (SELECT customer_key FROM cte2)
    GROUP BY customer_key) c ON a.customer_key = c.customer_key

    LEFT JOIN
    -- 最近一次申请的产品
    (SELECT DISTINCT customer_key, IF(COUNT(DISTINCT ord.product_type_key) = 1, ord.product_type_key, 3) '最近一次申请产品' FROM
        (SELECT customer_key, ord.product_type_key, DENSE_RANK() OVER (PARTITION BY customer_key ORDER BY ord.create_date_key DESC) 'order_rank' 
        FROM fact_crm_order ord 
        WHERE ord.product_type_key IN (1,2)
        AND ord.create_date_key < {exec_date_key}
        AND customer_key IN (SELECT customer_key FROM cte2))
    WHERE order_rank = 1
    GROUP BY customer_key) d ON a.customer_key = d.customer_key

    LEFT JOIN 

    -- 线上申请信用卡，信贷次数
    (SELECT DISTINCT customer_key, COUNT(DISTINCT IF(head_company_key = 35 AND product_type_key = 1, crm_order_key, NULL)) '线上申请信用卡次数', COUNT(DISTINCT IF(head_company_key = 35 AND product_type_key = 2, crm_order_key, NULL)) '线上申请信贷次数'
    FROM fact_crm_order ord
    WHERE ord.create_date_key < {exec_date_key}
    AND customer_key IN (SELECT customer_key FROM cte2)
    GROUP BY customer_key) e ON a.customer_key = e.customer_key

    LEFT JOIN 
    -- 距离最近一次申请信用卡的天数
    (SELECT DISTINCT customer_key, DATEDIFF({exec_date_key}, create_date_key) gap FROM
        (SELECT customer_key, create_date_key, DENSE_RANK() OVER (PARTITION BY customer_key ORDER BY create_date_key DESC) 'order_rank'
         FROM fact_crm_order ord
         WHERE product_type_key = 1
         AND ord.create_date_key < {exec_date_key}
         AND customer_key IN (SELECT customer_key FROM cte2))
    WHERE order_rank = 1 ) f ON a.customer_key = f.customer_key

    LEFT JOIN 
    -- 推送银行是否为曾经申请的银行
    (SELECT DISTINCT customer_key, IF(GROUP_CONCAT(card_name)  LIKE "%{bank}%", TRUE, FALSE) '是否曾经申请过相同银行',  GROUP_CONCAT(card_name) FROM fact_crm_order ord
    WHERE product_type_key = 1
    AND ord.create_date_key < {exec_date_key}
    AND customer_key IN (SELECT customer_key FROM cte2)
    GROUP BY customer_key) g ON a.customer_key = g.customer_key

    LEFT JOIN
    -- 推送银行是否为曾经核卡银行
    (SELECT DISTINCT customer_key, IFNULL(IF(GROUP_CONCAT(card_name)  LIKE "%{bank}%", TRUE, FALSE), FALSE) '是否有该银行的核卡',  GROUP_CONCAT(card_name) FROM fact_crm_order ord
    WHERE product_type_key = 1 AND order_status = "已核卡"
    AND ord.create_date_key < {exec_date_key}
    AND ord.audit_date_key < {exec_date_key}
    AND customer_key IN (SELECT customer_key FROM cte2)
    GROUP BY customer_key) h ON a.customer_key = h.customer_key

    LEFT JOIN 
    -- 推送银行是否为曾经拒绝银行
    (SELECT DISTINCT customer_key, IFNULL(IF(GROUP_CONCAT(card_name)  LIKE "%{bank}%", TRUE, FALSE), FALSE) '是否曾被该银行拒绝过',  GROUP_CONCAT(card_name) FROM fact_crm_order ord
    WHERE product_type_key = 1 AND order_status = "已拒绝"
    AND ord.create_date_key < {exec_date_key}
    AND ord.audit_date_key < {exec_date_key}
    AND customer_key IN (SELECT customer_key FROM cte2)
    GROUP BY customer_key) i ON a.customer_key = i.customer_key

    LEFT JOIN
    -- 推送银行是否和上一次申请的一致
    (SELECT DISTINCT customer_key, IF(GROUP_CONCAT(card_name) LIKE "%{bank}%", TRUE, FALSE) '是否和上一次申请的银行一致' FROM
        (SELECT customer_key, ord.card_name, DENSE_RANK() OVER (PARTITION BY customer_key ORDER BY ord.create_date_key DESC) 'order_rank' 
        FROM fact_crm_order ord 
        WHERE ord.product_type_key = 1
        AND ord.create_date_key < {exec_date_key}
        AND customer_key IN (SELECT customer_key FROM cte2))
    WHERE order_rank = 1
    GROUP BY customer_key) j ON a.customer_key = j.customer_key
    
    LEFT JOIN 
    -- 客户近一个月推送次数
    (SELECT customer_key, COUNT(1) '近一月推送次数' FROM fact_dmp_send_record_detail detail
    WHERE send_status_key = 2
    AND customer_key IN (SELECT customer_key FROM cte2)
    AND detail.send_date_key BETWEEN CAST(DATE_FORMAT(SUBDATE(20220309, INTERVAL 1 MONTH), "%Y%m%d") AS DECIMAL(20, 0)) AND 20220309
    GROUP BY customer_key) k ON a.customer_key = k.customer_key

    )
    '''
#     print(collect_train_data)
    try:
        connection = create_db_connection("am-wz98vqjuhh95217b1167320o.ads.aliyuncs.com","dmc_ads_read","yumbyb3JspVUvTHk", "dmc_abs_dw")
        train_data = pd.DataFrame(read_query(connection, collect_train_data))
        train_data.columns = ['customer_key', '下单用户', 'customer_key.1', 'age', 'gender', 'mob',
           '信用卡申请数', '信贷申请数', '核卡数', '被拒数', '核卡率', '被拒率', '最近一次申请产品', '线上申请信用卡次数',
           '线上申请信贷次数', '距离上一次申请信用卡天数', '是否曾经申请过相同银行', '是否有该银行的核卡', '是否曾被该银行拒绝过',
           '是否和上一次申请的银行一致']
        train_data.to_csv(f"{outpath}/{exec_date_key}-{wechat_account}-{bank}-{job_name}.csv", index=False)
    except:
        print("===========this fail==============")
        fail_job.append(job_key)
#     break
#     train_datas.append(pd.DataFrame(read_query(connection, collect_train_data)))
```
